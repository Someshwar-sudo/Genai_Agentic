# TEXT-CLEANING-PROCESS

## Both Stemming and Lemmatization are techniques used to clean text by reducing words down to their root or base form. This helps machine learning models realize that words like "running", "runs", and "ran" all mean the same core action

## step-by-step order of how your NLP code actually executes the text cleaning process.

#### Step 1: Import the packages


```python
import re
import nltk
from nltk.corpus import stopwords   #stopwords
from nltk.stem import WordNetLemmatizer # lematizer
from nltk.stem.porter import PorterStemmer
```

#### Step 2: Initialize the tools and storage
##### You create the objects for stemming/lemmatization and create an empty list (corpus) to store the cleaned sentences later.

```python
ps = PorterStemmer()
wordnet = WordNetLemmatizer()
corpus = []
```

#### Step 3: Tokenize the paragraph into sentences
##### Before cleaning individual words, you must split the giant block of text into a list of single sentences.

```python
sentences='Hi world !. How are you !!. and whats up!!'
sentences = nltk.sent_tokenize(sentences)
print('senteces :',sentences)
print('length of sentances :',len(sentences))
```

#### Note: Steps 4 through 9 happen inside a for loop, processing one sentence at a time.

#### Step 4: Remove special characters and numbers
#####  use a regular expression to find anything that is not a letter (like punctuation, numbers, or symbols) and replace it with a blank space.

```python
review = re.sub("[^a-zA-Z\s]", "", sentences)
```

#### Step 5: Convert text to lowercase
#### make all letters lowercase so the computer does not treat "The" and "the" as two completely different words.

```python
review = review.lower()
```

#### Step 6: Split the sentence into individual words
##### break the single string sentence into a list of individual words. This is necessary because the lemmatizer and stopword filters only work on individual words

```python
review = review.split()
```

#### Step 7: Remove stopwords
##### check every word against NLTK's English stopword list and throw away filler words like "the", "is", and "at" because they do not carry unique meaning.

```python
# This happens inside the list comprehension:
# if not word in set(stopwords.words('english'))
# refer step-8
```

#### Step 8: Apply Lemmatization (or Stemming)
##### While filtering out the stopwords simultaneously convert the remaining words down to their root dictionary form using the lemmatizer.

```python
review = [
    wordnet.lemmatize(word)
    for word in review
    if not word in set(stopwords.words("english"))
]
```

#### Step 9: Join the words back into a sentence
##### take your cleaned list of words and glue them back together with spaces to form a clean, readable sentence string.

```python
review = " ".join(review)
```

#### Step 10: Append to the corpus list
##### You save the final, fully cleaned sentence string into your corpus storage list.

```python
corpus.append(review)
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import sent_tokenize


wordnet=WordNetLemmatizer()
Ps=PorterStemmer()

paragraph = """Virat Kohli is a famous Indian cricketer. He has broken many world records with his bats. 
               Fans love watching his aggressive style on the stadium fields! He always plays for his nation."""

sentences=sent_tokenize(paragraph)

corpus=[]

for i in range (len(sentences)):
    review=re.sub('[^a-zA-Z]', ' ', sentences[i]) # removing special characters
    review=review.lower()        #making every letter in small letters
    review=review.split()        # make sentence to words 
    review=[wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))] #apply lammentatision
    review=''.join(review)
    corpus.append(review)
    
    
# crete the bow model 
from sklearn.feature_extraction.text import CountVectorizer
cv1 = CountVectorizer()
X1 = cv1.fit_transform(corpus).toarray()   
print(X1)
    
    
# Creating the TF-IDF model
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
X = cv.fit_transform(corpus).toarray()
print(X)


    
    


```
