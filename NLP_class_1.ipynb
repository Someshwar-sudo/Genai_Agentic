### Natural Language Processing (NLP) is a technology that trains computers to understand, interpret, and generate human language by breaking down text and speech into structured data

### NLP WORK FLOW
### RAW TEXT ==>TOKENIZATION==>REMOVE STOP WORDS ==>Stemming/Lemmatization==>vectorization==>Machinelearn model



### 1.Tokenization:- senetence splits into individual words or units 
### Example :- 'i love india' ==> ['i ','love','india']


### 2.Stop word removal
### Filter out comman words like 'and' ,'is','or'

### 3.Stemming/lematization (ing to base)
### Reduces words to their base or dictionary form so the system understands they mean the same thing
### example :- running converts to run 

### 4.Vectorization (Word Embeddings)
### Converts the cleaned text tokens into mathematical numbers (vectors) because computers only understand math. Words with similar meanings are placed close together in a multi-dimensional mathematical space

### 5.Model ProcessingAction: 
### Feeds these numbers into machine learning models (like Transformers or Neural Networks) to predict intent, generate a response, or classify the sentiment.

### Token Frequency

```python
# lets count the frequncy of words
from nltk.probability import FreqDist
FreqDist=FreqDist(words)
FreqDist
```

### lets make all captial letters into small letters

```python
for i in words:
    FreqDist[i.lower()]+=1
FreqDist
```

### Types of tokens

### Bigrams, trigrams, and n-grams are contiguous sequences of \(n\) items (words or characters) sliced from a given sample of text. They are useful because they preserve local context and word order, which single-word tokens (unigrams) completely lose

### In sentiment analysis, the phrase "not good" broken into unigrams yields ["not", "good"]. The machine flags "good" as positive, misinterpreting the overall meaning.

```python
from nltk import bigrams
print(bigrams.__module__)
```

```python
from nltk.util import bigrams,trigrams,ngrams
bigrams_token=list(bigrams(words))
bigrams_token
```

```python
trigrams=list(trigrams(words))
trigrams
```

```python
ngrams=list(ngrams(words,4))
ngrams
```
