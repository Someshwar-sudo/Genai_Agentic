```python
### Question - Find the longest token in the sentence. Machine Learning is amazing
text2='Machine learning is amazing'
token2=text2.split()
print(token2)
largest_token=max(token2,key=len)
largest_token
```

```python
### Question - Find the shortest token.
text2='Machine learning is amazing'
token2=text2.split()
print(token2)
shortest_token=min(token2,key=len)
print(shortest_token)
```

```python
## Question - Create vocabulary from text.
text = "I love AI and NLP"
tokens = text.split()
print(tokens)
vocab = sorted(set(tokens))
print(vocab)
```

```python
## Question - Convert tokens into token IDs
for i ,word in enumerate(vocab):
    print(i,word)
for word in tokens:
    print(vocab[word])
```

```python
text = "I love AI"
tokens = text.split()
vocab = {word:i for i,word in enumerate(set(tokens))}
token_ids = [vocab[word] for word in tokens]
print(token_ids)
```

```python
# Question - Decode token IDs back into text.
# Answer:-
vocab = {
 "I":0,
 "love":1,
 "AI":2
}
reverse_vocab = {v:k for k,v in vocab.items()}
ids = [0,1,2]
sentence = [reverse_vocab[i] for i in ids]
print(" ".join(sentence))
```

```python
# Question - Implement a simple tokenizer class.
# Answer:-
class SimpleTokenizer:
    def tokenize(self,text):
        return text.split()
obj = SimpleTokenizer()
print(obj.tokenize("I love AI"))
```
